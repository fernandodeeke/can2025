```python
#Brusselator

import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import solve_ivp
import ipywidgets as widgets
from ipywidgets import interact

# Fixed parameters
T = 100
x0 = 1.5
y0 = 1.0
IC = [x0, y0]

# Time grid
time = np.linspace(0, T, 1000)

# System
def sistema(t, S, a, b):
    x, y = S
    dxdt = a - (b + 1) * x + (x**2) * y / 4
    dydt = b * x - (x**2) * y / 4
    return [dxdt, dydt]

# Interactive plotting function
def plot_brusselator(a=1.0, b=1.0):
    sol = solve_ivp(
        lambda t, S: sistema(t, S, a, b),
        [0, T],
        IC,
        t_eval=time,
        method="RK45"
    )

    x, y = sol.y

    fig, axes = plt.subplots(1, 2, figsize=(12, 4), dpi=120)

    # Time series
    axes[0].plot(time, x, label="x", lw=1.5)
    axes[0].plot(time, y, label="y", lw=1.5)
    axes[0].set_title(f"Time series (a={a:.2f}, b={b:.2f})")
    axes[0].set_xlabel("$t$")
    axes[0].set_ylabel("$x$, $y$")
    axes[0].grid(True)
    axes[0].legend()

    # Phase portrait
    axes[1].plot(x, y, lw=1.5)
    axes[1].plot(x[0], y[0], 'o', label='initial point')
    axes[1].plot(x[-1], y[-1], 's', label='final point')
    axes[1].set_title("Phase portrait")
    axes[1].set_xlabel("$x$")
    axes[1].set_ylabel("$y$")
    axes[1].grid(True)
    axes[1].legend()

    plt.tight_layout()
    plt.show()

# Sliders
interact(
    plot_brusselator,
    a=widgets.FloatSlider(value=1.0, min=0.2, max=3.0, step=0.1, description='a'),
    b=widgets.FloatSlider(value=1.3, min=0.1, max=8.0, step=0.1, description='b')
);
```
