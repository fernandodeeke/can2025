```python
from sympy import *
```

```python
from sympy.plotting import plot 
```

```python
x= symbols('x')
f = Lambda(x,1/(x**2+cos(x)+4))
f
```

```python
plot( f(x), (x, 0, 1))
```

```python
g = Lambda(x,cos(x**2))
g
```

```python
g(x).series(x,0,n=10)
```

```python
g_taylor = Lambda(x,1-x**4/2+x**8/24)
g_taylor
```

```python
plot( g(x), g_taylor(x), (x, 0, 1.5),line_color='red')
```
