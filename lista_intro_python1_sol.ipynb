2. (a) Plote o  gráfico de $y= \sqrt{x} \sin(60 x)$ no intervalo de  $x = 0$ a $x = 3\pi$.<br>
(b) Plote as curvas $y_1= \sqrt{x} \sin(60 x)$, $y_2=\sqrt{x}$, $y_3=-\sqrt{x}$ simultaneamente e mostre que $y_1$ está "presa" entre $y_2$ e $y_3$. 

Solução. Definimos as funções:

```python
 def y1(x):
        return sqrt(x)*sin(60*x)
```

```python
 def y2(x):
        return sqrt(x)
```

```python
def y3(x):
        return -sqrt(x)
```

```python
x=linspace(0,1,300)
fig, ax = subplots()
ax.grid(True, which='both')
ax.plot(x, y1(x),color='blue')
ax.plot(x, y2(x),color='red')
ax.plot(x, y3(x), color='red')
ax.axhline(y=0, color='black')
ax.axvline(x=0, color='black')
xlabel('x')
ylabel('y')
title('Modulação da amplitude')
legend(['Função periódica', 'Envelope'])
```

Este gráfico mostra que $y_2$ e $y_3$ são envelopes da curva $y_1$.

8. Determine as raízes reais da equação
$$\sin(x^2) +\sin(x)-x^2+3=0$$
e ilustre graficamente:

Solução. Façamos inicialmente o gráfico: 

```python
 def f(x):
        return sin(x**2)+sin(x)-x+3
```

```python
x=linspace(-5,5,100)
fig, ax = subplots()
ax.grid(True, which='both')
ax.plot(x, f(x))
ax.axhline(y=0, color='black')
ax.axvline(x=0, color='black')
```

O gráfico mostra que temos somente uma raiz real, no intervalo [2,4]. Para resolver equações algébricas não lineares no Python, usaremos o solver presente no pacote scipy.optimize. O método utilizado é o de [Broyden](https://en.wikipedia.org/wiki/Broyden%27s_method), que é uma generalização dos métodos de Newton-Raphson e secantes que inclui sistemas de equações. Usemos o comando fsolve e um ponto inicial próximo à raiz, digamos $x=3$:

```python
rr=fsolve(f,3)
r=rr[0]
r
```

Podemos verificar que esta é de fato uma raiz: 

```python
f(r)
```
