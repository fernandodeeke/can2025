## Método de Newton-Raphson

Devemos implementar a fórmula de Newton-Raphson:
$$
x_{i+1} = x_i-\frac{f(x_i)}{f'(x_i)}\,.
$$

```python
import numpy as np
import matplotlib.pyplot as plt
from sympy import *
```

```python
x = symbols('x')
```

```python
f = Lambda(x, x**2 - 3*cos(x**2)+x -3)
```

```python
plot(f(x), (x, -4,5))
```

Determinemo o zero positivo de $f$.

```python
df = diff(f(x), x)
df
```

```python
Df = Lambda(x,df)
Df(x)
```

```python
x0 = -2.1
for i in range(1,5):
  x1 = x0-f(x0)/Df(x0)
  x0 = x1
  print(f'x{i} = {x1}')
```

```python
f(x1)
```

```python
def newton_raphson(f,x0,N):
  for i in range(1,5):
    x1 = x0-f(x0)/Df(x0)
    x0 = x1
  return x0
```
